```python
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
```

```python
def classify_dataframe_in_batches(model, tokenizer, input_df, label_mapping, batch_size=32):
    result_df = pd.DataFrame(columns=['text', 'label'])
    model = model.to(device)
    
    for i in range(0, len(input_df), batch_size):
        batch_df = input_df.iloc[i:i+batch_size]
        

        inputs = tokenizer(list(batch_df['text']), return_tensors="pt", padding=True, truncation=True, max_length=512).to(device)
        outputs = model(**inputs)

        probabilities = torch.nn.functional.softmax(outputs.logits, dim=-1)
        predicted_classes = torch.argmax(probabilities, dim=1).tolist()
        predicted_classes = list(map(label_mapping.get, predicted_classes))

        batch_result_df = pd.DataFrame({'text': batch_df['text'].tolist(), 'label': predicted_classes})
        result_df = pd.concat([result_df, batch_result_df], ignore_index=True)

    return result_df
```

```python
model_large_path = '/Users/bellesbae/ml_torch/hacks-2/rubert_large_73'
model_small_path = "/Users/bellesbae/ml_torch/hacks-2/bert_small_labels"
tokenizer_path = "/Users/bellesbae/ml_torch/hacks-2/bert_tokenizer"

tokenizer = AutoTokenizer.from_pretrained(tokenizer_path)
model_large = AutoModelForSequenceClassification.from_pretrained(model_large_path)
model_small = AutoModelForSequenceClassification.from_pretrained(model_small_path)
```

### Test big labels

```python
test_data = pd.read_csv("/Users/bellesbae/ml_torch/hacks-2/test.csv", sep=';')
test_large_data = test_data[['Текст инцидента']]
test_large_data.columns = ['text']
```

```python
result_large = classify_dataframe_in_batches(model_large, tokenizer, test_large_data, id2label_large, batch_size=128)
result_large
```

### Test of small labels

```python
test_small_data = pd.DataFrame()
test_small_data['text'] = result_large['label'] + ": " + test_data['Текст инцидента']
result_small = classify_dataframe_in_batches(model_small, tokenizer, test_small_data, id2label_small, batch_size=128)
result_small
```

```python
submission = pd.DataFrame(data={
    'id': result_small.index,
    "Группа тем": result_large['label'],
    "Тема": result_small['label'],
})
submission
```
